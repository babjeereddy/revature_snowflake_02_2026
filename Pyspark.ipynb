```python
# groupByKey() is a transformation in PySpark RDD that groups values with the same key.

# It works only on Pair RDDs (key-value RDD).
```

```python
reduceByKey()
    Aggregates values immediately
    Uses a function like sum, max, min
    Returns one value per key

groupByKey()
    Groups all values into a list
    Returns Iterable of values per key
    Does NOT perform aggregation automatically

reduceByKey()
    Performs local aggregation inside each partition
    Then shuffles only reduced data
    Less network traffic
    Faster
groupByKey()
    Shuffles ALL values across network
    Then groups them
    High memory usage
    Slower

Feature	reduceByKey	groupByKey
Shuffle Volume	Low	High
Memory Usage	Low	High
Speed	Fast	Slow
Returns	(K, V)	(K, Iterable[V])
```

```python
# groupByKey
# For each product, find its aggregated revenue.
ordItemFilter= ordItems.filter(lambda x : int( x.split(',')[1] ) <501)
ordGrp = ordItemFilter.map(lambda x : (int(x.split(',')[1]),float(x.split(',')[4]))).groupByKey()
result = ordGrp.mapValues(sum).take(10)
for x in result:
    print(x)
```
